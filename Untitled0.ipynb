```python
target_list = [
    'https://www.sandbox.game/en/map/?liteMap=false&currentX=-1961&currentY=-832&zoom=6.5&sizes=4'
]

test_header={
    'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Cache-Control':'max-age=0',
    'Connection':'keep-alive',
    'Referer':'https://courses.students.ubc.ca/',
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.104 Safari/537.36 Core/1.53.4882.400 QQBrowser/9.7.13059.400'
    # notice that the user agent may vary depending on the platform
}
```

```python
def find_general_information(soup):
    tag = soup.find(text = re.compile("Head"))
    return tag
```

```python
for target in target_list:
    num = random.random()*2+1
    time.sleep(num)
    # print(num)  
    # res=requests.get(url,headers=headers)
    url = target
    try:
        html = requests.get(url,headers = test_header)
        soup = BeautifulSoup(html.text,'lxml')
        print(find_general_information(soup))
    except:
        print('the web page is not giving a return')
        time.sleep(1)
```
